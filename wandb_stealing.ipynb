```python
df1 = pd.read_csv('wandb_data/good_amp_tloss.csv')
print(df1.columns)
t_acc = df1['noisy_bin_precision - train_loss']
t_acc = t_acc.to_numpy()

df2 = pd.read_csv('wandb_data/good_amp_eloss.csv')
v_acc = df2['noisy_bin_precision - eval_loss']
v_acc = v_acc.to_numpy() # offset by 5 epochs

df3 = pd.read_csv('wandb_data/good_amp_tacc.csv')
t_loss = df3['noisy_bin_precision - train_acc']
t_loss = t_loss.to_numpy()

df4 = pd.read_csv('wandb_data/good_amp_eacc.csv')
v_loss = df4['noisy_bin_precision - eval_acc']
v_loss = v_loss.to_numpy() # offset by 5 epochs
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Epoch indices
training_epochs = np.arange(len(t_acc))
val_epochs = np.arange(len(v_acc)) * 5

# training_loss = np.arange(len(t_loss)) * 5  # validated every 5 epochs
# val_loss = np.arange(len(v_loss)) * 5

# Create plots
fig, axs = plt.subplots(1, 2, figsize=(12, 5))

# --- Training Accuracy ---
axs[0].plot(training_epochs, t_loss, label='Training Acc', color='tab:blue', marker='o')
axs[0].plot(val_epochs + 5, v_loss, label='Val Acc', color='tab:orange', marker='o')
axs[0].set_title('Accuracy')
axs[0].set_xlabel('Epoch')
axs[0].set_ylabel('Classification Accuracy')
axs[0].legend()
axs[0].grid(True)

# --- Validation Accuracy ---
axs[1].plot(training_epochs, t_acc, label='Training Loss', marker='o', color='tab:blue')
axs[1].plot(val_epochs + 5, v_acc, label='Val Loss', marker='o', color='tab:orange')
axs[1].set_title('Loss')
axs[1].set_xlabel('Epoch')
axs[1].set_ylabel('Sampled + Weighted BCE Loss')
axs[1].legend()
axs[1].grid(True)

plt.tight_layout()
plt.show()
```

```python
df5 = pd.read_csv('wandb_data/trainacc_trans_1us.csv')
trans_ta2 = df5['loss-development - train_acc']
trans_ta2 = trans_ta2.to_numpy()

df6 = pd.read_csv('wandb_data/valacc_trans_1us.csv')
trans_va2 = df6['loss-development - eval_acc']
trans_va2 = trans_va2.to_numpy() # offset by 5 epochs

df7 = pd.read_csv('wandb_data/trainacc_unet_1us.csv')
unet_ta2 = df7['loss-development - train_acc']
unet_ta2 = unet_ta2.to_numpy()
```

```python
# Epoch indices
trans_ta2_epochs = np.arange(len(trans_ta2))
unet_ta2_epochs = np.arange(len(unet_ta2))

trans_va2_epochs = np.arange(len(trans_va2)) * 5  # validated every 5 epochs
# unet_va_epochs = np.arange(len(unet_va)) * 5

# Create plots
fig, axs = plt.subplots(1, 2, figsize=(12, 5))

# --- Training Accuracy ---
axs[0].plot(trans_ta2_epochs, trans_ta2, label='Transformer', color='tab:blue', marker='o')
axs[0].plot(unet_ta2_epochs, unet_ta2, label='UNet', color='tab:orange', marker='o')
axs[0].set_title('Training Accuracy')
axs[0].set_xlabel('Epoch')
axs[0].set_ylabel('Accuracy')
axs[0].legend()
axs[0].grid(True)

# --- Validation Accuracy ---
axs[1].plot(trans_va2_epochs + 5, trans_va2, label='Transformer', marker='o', color='tab:blue')
axs[1].set_title('Validation Accuracy')
axs[1].set_xlabel('Epoch')
axs[1].set_ylabel('Accuracy')
axs[1].legend()
axs[1].grid(True)

plt.tight_layout()
plt.show()
```
