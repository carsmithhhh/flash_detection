### Generating Signals Every 100 Picoseconds

```python
timing_offsets = torch.linspace(498, 509, 11) # units of 1/10 ns
print(timing_offsets)

pmt_ids = torch.full((1,), 0)

peak_signals = []
peak_bins = []
arrival_signals = []
arrival_bins = []

threshold = -1e-4

for i, val in enumerate(timing_offsets):
    # generate photon timing distribution
    arrival_times = torch.zeros(1, dtype=torch.int64) + int(val.item())
    wf = sim.gen_waveform(mode='precise', pmt_ids=pmt_ids, arrival_times=arrival_times, n_pmts=1)
    
    # convolve with detector response
    output, info = sim(wf.to(device), scintillation=False, tpb_delay=False, combined=False)
    output = output.detach().cpu().numpy()  # shape [1000]
    
    # check arrival bin (first nonzero bin should be ~50 ns after offset)
    above_thresh = np.where(output < threshold)[0]
    if len(above_thresh) == 0:
        print(f"No signal detected for offset {val.item()}")
        continue

    arrival_bin = above_thresh[0]
    arrival_bins.append(arrival_bin)
    arrival_signals.append(output[arrival_bin])
    # print(output[int(val.item())], output[int(val.item())+1])
    # arrival_bin = int(val.item() / sim.downsample_factor)
    # arrival_bins.append(arrival_bin)
    # arrival_signals.append(output[arrival_bin])
    
    # get peak signal (largest value in the waveform)
    peak_bin = np.argmin(output)
    peak_bins.append(peak_bin)
    peak_signals.append(output[peak_bin])    

print(arrival_bins)
plt.plot(output)
plt.grid()
plt.axvline(x=arrival_bin, color='r', linestyle='--', linewidth=2, label='True arrival')
```

```python
ticks = np.linspace(0, 10, 11)
peak_signals = np.asarray(peak_signals)
arrival_signals = np.asarray(arrival_signals)
peak_bins = np.asarray(peak_bins)
arrival_bins = np.asarray(arrival_bins)

print("arrival_bins:", arrival_bins)

fig, axes = plt.subplots(1, 2, figsize=(12, 4))

# Left: signal amplitudes
axes[0].plot(ticks, peak_signals, label='peak value', marker='o')
axes[0].plot(ticks, arrival_signals, label='arrival bin value', marker='o')
axes[0].set_xlabel("Arrival Time Within 1 Bin (1/10 ns)")
axes[0].set_ylabel("ADC")
axes[0].set_xticks(ticks)  # Positions (unchanged)
axes[0].set_xticklabels([f"{t/10:.1f}" for t in timing_offsets])  # Custom labels, e.g., convert to ns
axes[0].set_title("Effect of Sub-Bin Precision on Peak & Arrival Bin Signals")
axes[0].grid()
axes[0].legend()

# Right: bin indices
axes[1].plot(ticks, peak_bins, label='peak bins', marker='o')
axes[1].plot(ticks, arrival_bins, label='arrival bins', marker='o')
axes[1].set_xlabel("Arrival Time Within 1 Bin (1/10 ns)")
axes[1].set_xticks(ticks)  # Positions (unchanged)
axes[1].set_xticklabels([f"{t/10:.1f}" for t in timing_offsets])  # Custom labels, e.g., convert to ns
axes[1].set_ylabel("Bin Index (ns)")
axes[1].set_title("Peak & Arrival Bins")
axes[1].legend()

plt.tight_layout()
plt.grid()
plt.show()
```

### Generating Data for Bin Precision Study (no noise)
**Uniform Distribution**
- 100,000 flashes, with random photon arrival times (floating point)
- Each flash has between 1 and 1,000 photons

**Every 100 Picoseconds**
- 100,000 flashes, with 10 photon arrival times at 100 picosecond intervals
- Each 100 picosecond interval has flashes with np.linspace(1, 1000, 10) photons

```python
# Uniform
nphotons = torch.linspace(1, 1000, 20).round().to(torch.int64)
arrival_times = torch.arange(0, 6000, 10)
sim_waveforms = []
all_arrival_times = []

for arrival_time in arrival_times:
    for n in nphotons:
        n_int = n.item()
        t_int = arrival_time.item()

        pmt_ids = torch.zeros(n_int, dtype=torch.int64)
        arrival_times_tensor = torch.full((n_int,), t_int, dtype=torch.int64)

        wf = sim.gen_waveform(mode='precise', pmt_ids=pmt_ids, arrival_times=arrival_times_tensor, n_pmts=1)
        output, info = sim(wf.to(device), scintillation=False, tpb_delay=False, combined=False)
        output = output.detach().cpu().numpy()

        sim_waveforms.append(output)
        all_arrival_times.append(t_int)  # store raw arrival time

all_arrival_times = np.array(all_arrival_times)
arrival_times_scaled = (all_arrival_times / sim.downsample_factor) + (offset / sim.downsample_factor)

data = {
    'waveforms': sim_waveforms,                      # list of NumPy arrays
    'arrival_times': arrival_times_scaled            # NumPy array
}

np.save("uniform_12k.npy", data, allow_pickle=True)
```

### Analyzing Model Performance

```python
# --- Set seed for reproducibility ---
seed = 42
random.seed(seed)
np.random.seed(seed)
torch.manual_seed(seed)
torch.cuda.manual_seed_all(seed)

# Loading in data
load_wfs = np.load('even_sample_2phot_noisy_100k.npy', allow_pickle=True) # array of shape (2519, 5, 250, 250)
dataset = WaveformDataset(load_wfs.item())

g = torch.Generator()
g.manual_seed(seed)

# Splitting data
val_ratio = 0.0
test_ratio = 0.0
total_size = len(dataset)
val_size = int(total_size * val_ratio)
test_size = int(total_size * test_ratio)
train_size = total_size - val_size - test_size
print(f"split sizes: {train_size, val_size, test_size}")
train_dataset, val_dataset, test_dataset = random_split(dataset, [train_size, val_size, test_size], generator=g)

batch_size = 25

def custom_collate_fn(batch):
    """
    Custom collate function for WaveformDataset.
    Each item in batch is a tuple: (waveform, arrival_time).
    Returns:
        waveforms: Tensor of shape (batch_size, waveform_length)
        arrival_times: Tensor of shape (batch_size,) or (batch_size, 1)
        hit_times: Tensor of shape (?) with a list of hit times per sample
    """
    waveforms, arrival_times, hit_times = zip(*batch)
    waveforms = torch.stack(waveforms, dim=0)
    waveforms = waveforms.unsqueeze(1)  # adding channel dimension

    # for binary classification
    arrival_times = torch.stack(arrival_times, dim=0)
    arrival_times = arrival_times.unsqueeze(1) # adding channel dimension

    # for regression, just use hit times
    hit_times = [item[2] for item in batch]
    hit_times = torch.tensor(hit_times)
    
    return waveforms, arrival_times, hit_times
    
train_loader = DataLoader(
    train_dataset,
    batch_size=batch_size,
    shuffle=True,
    generator=g,
    collate_fn=custom_collate_fn,
    num_workers=0,
    pin_memory=False,
    drop_last=False
)

val_loader = DataLoader(
    val_dataset,
    batch_size=batch_size,
    shuffle=False,
    collate_fn=custom_collate_fn,
    num_workers=0,
    pin_memory=False,
    drop_last=False
)

test_loader = DataLoader(
    test_dataset,
    batch_size=batch_size,
    shuffle=False,
    collate_fn=custom_collate_fn,
    num_workers=0,
    pin_memory=False,
    drop_last=False
)
```

### Accuracy by Sub-Bin Timing

```python
# --- Load model ---
from model import *
model = UNet1D()
checkpoint = torch.load("correct_amp_unet50.pth", weights_only=True)
model.load_state_dict(checkpoint['model_state_dict'])
model = model.to('cuda')  # move model to GPU if needed
model.eval()

results = np.zeros(10)
totals = np.zeros(10)

val_progress = tqdm(train_loader, leave=False, position=0)
model.eval()

deltas = []

with torch.no_grad():
    for val_data, val_target, val_hit_times in val_progress:
        val_data = val_data.to(device)  # shape: [B, 1, 1000]
        val_output = model(val_data, mode='bce')  # shape: [B, 1, 1000]
        val_pred = torch.sigmoid(val_output).squeeze(1)  # shape: [B, 1000]
        for b in range(val_pred.shape[0]):
            true_hits = val_hit_times[b]
            if isinstance(true_hits, torch.Tensor):
                true_hits = true_hits.cpu().tolist()
            if len(true_hits) == 0:
                continue  # Skip if no true hits
    
            # Shifted hit indices for bin calculation
            real_true_hits = [(x + 100) for x in true_hits]
            shifted_hits = [int(round(x + 100)) for x in true_hits]
            # Update totals by fractional bin (last digit of bin index)
            for hit in real_true_hits:
                last_digit = int((hit * 10) % 10)
                totals[last_digit] += 1
                 # Only when ovelapping on purpose !!!!!!!!!!!!!!
    
            # Predict top-k bin indices
            topk_pred_indices = torch.topk(val_pred[b], k=len(true_hits)).indices.cpu().numpy()
            # topk_pred_indices = torch.topk(val_pred[b], k=25).indices.cpu().numpy() # Only when ovelapping on purpose !!!!!!!!!!!!!!
            # Match predictions to ground truth hits
            for p in topk_pred_indices:
                # print(p)
                matching_indices = [i for i, t in enumerate(shifted_hits) if t == p+2]
    
                for idx in matching_indices:
                    last_digit = int(real_true_hits[idx] * 10 % 10)
                    # print(last_digit)
                    results[last_digit] += 1
                # print(p, real_true_hits[0])
                deltas.append(p - real_true_hits[0])
            
                    # Only when ovelapping on purpose !!!!!!!!!!!!!!
    # for val_data, val_target, val_hit_times in val_progress:
    #     val_data = val_data.to(device)       # shape: [B, 1, 1000]
    #     val_output = model(val_data, mode='bce')  # shape: [B, 1, 1000]
    #     val_pred = torch.sigmoid(val_output).squeeze(1)  # [B, 1000]
        
    #     # val_hit_times: list of length B, each item is list of ints
    #     for b in range(val_pred.shape[0]):
            
    #         # True hit times for this sample
    #         if len(true_hits) == 0:
    #             continue  # skip if no true hits
                
    #         true_hits = val_hit_times[b]
    #         if isinstance(true_hits, torch.Tensor):
    #             true_hits = true_hits.cpu().tolist()
                
    #         real_true_hits = [(x+100) for x in true_hits]
    #         true_hits = [int(x+100) for x in true_hits]

    #         # Build total # of true flashes per bin
    #         for i in range(len(true_hits)):
    #             true_idx = true_hits[i]  
    #             true_last_digit = int((true_idx * 10) % 10)
    #             totals[true_last_digit] += 1

    #         # Get top k predictions
    #         topk_pred_indices = torch.topk(val_pred[i], k=len(true_hits)).indices.detach().cpu().numpy()
            
    #         # pred_idx = torch.argmax(val_pred[b]).item()  # predicted arrival index
            
    #         for p in topk_pred_indices:
    #             correct_idx = np.where(true_hits == p)[0]
    #         print(p)
    #         print(true_hits)
    #         print(correct_idx)
    #         break
            
    #         for idx in correct_idx:
    #             true_last_digit = int(real_true_hits[idx] * 10 % 10)
    #             results[true_last_digit] += 1

    #         deltas.append(true_hits[0] - pred_idx)
    #     break

# Compute accuracies safely
accuracies = np.divide(results, totals, out=np.zeros_like(results), where=totals > 0)

# Print results
print("Accuracy by last digit of predicted arrival time (0–9):")
for i in range(10):
    print(f"Last digit {i}: {accuracies[i]:.3f} ({int(results[i])}/{int(totals[i])})")
```

### Accuracy by Signal Amplitude

```python
# --- Load model ---
from collections import defaultdict

# model = UNet1D()
# checkpoint = torch.load("unet_noisy_bin_precision_110epochs.pth", weights_only=True)
# model.load_state_dict(checkpoint['model_state_dict'])
# model = model.to('cuda')  # move model to GPU if needed
# model.eval()

# Set bin edges: from 0 to, say, 1000 in steps of 50 units
bin_edges = np.arange(0, 100, 10)  # You can increase the upper bound if needed
bin_edges = np.append(bin_edges, 300)
num_bins = len(bin_edges) - 1

# Initialize result counters
totals = np.zeros(num_bins, dtype=int)
corrects = np.zeros(num_bins, dtype=int)

val_progress = tqdm(val_loader, leave=False, position=0)

with torch.no_grad():
    for val_data, val_target, val_hit_times in val_progress:
        val_data = val_data.to(device)       # shape: [B, 1, 1000]
        val_output = model(val_data, mode='bce')  # shape: [B, 1, 1000]
        val_pred = torch.sigmoid(val_output).squeeze(1)  # [B, 1000]
        
        for b in range(val_pred.shape[0]):
            pred_idx = torch.argmax(val_pred[b]).item()  # predicted arrival index
        
            true_hits = val_hit_times[b]
            if isinstance(true_hits, torch.Tensor):
                true_hits = true_hits.cpu().tolist()

            if len(true_hits) == 0:
                continue

            true_idx = int(true_hits[0])  # extract as Python int
            start_idx = max(true_idx - 50, 0)
            end_idx = min(true_idx + 100, val_output.shape[2])
            window = val_data[b, 0, int(start_idx):int(end_idx)]
            true_amp = torch.abs(torch.min(window)).item()

            # Determine bin index
            bin_idx = np.digitize(true_amp, bin_edges) - 1  # digitize returns 1-based indices
            if 0 <= bin_idx < num_bins:
                totals[bin_idx] += 1

                # Shift hits for prediction comparison
                true_hits_shifted = [int(x + 100) for x in true_hits]
                if pred_idx in true_hits_shifted:
                    corrects[bin_idx] += 1

# Calculate accuracy
accuracies = np.divide(corrects, totals, out=np.zeros_like(corrects, dtype=float), where=totals > 0)

# Print results
print("Accuracy by Signal Amplitude Range:")
for i in range(num_bins):
    left = bin_edges[i]
    right = bin_edges[i+1]
    print(f"{left:4.0f}–{right:4.0f}: {accuracies[i]:.3f} ({corrects[i]}/{totals[i]})")
```
