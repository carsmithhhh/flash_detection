```python
import sys
sys.path.append('..')
from data_utils import *

import numpy as np
import matplotlib.pyplot as plt
import pickle

import torch
from torch.utils.data import DataLoader, random_split
from torch.utils.data import Subset
import torch.optim as optim
from torch.optim.lr_scheduler import SequentialLR, LinearLR, CosineAnnealingLR

device = 'cuda'
```

```python
# Loading in data
load_wfs = np.load('10k_flashes.npy', allow_pickle=True) # array of shape (2519, 5, 250, 250)
print(load_wfs.item()['arrival_times'])
dataset = WaveformDataset(load_wfs.item())

batch_size = 25

def custom_collate_fn(batch):
    """
    Custom collate function for WaveformDataset.
    Each item in batch is a tuple: (waveform, arrival_time).
    Returns:
        waveforms: Tensor of shape (batch_size, waveform_length)
        arrival_times: Tensor of shape (batch_size,) or (batch_size, 1)
    """
    waveforms, arrival_times = zip(*batch)
    waveforms = torch.stack(waveforms, dim=0)
    waveforms = waveforms.unsqueeze(1).to(device)  # adding channel dimension

    arrival_times = torch.stack(arrival_times, dim=0)
    arrival_times = arrival_times.unsqueeze(1).to(device) # adding channel dimension
    return waveforms, arrival_times

loader = DataLoader(
    dataset,
    batch_size=batch_size,
    shuffle=False,
    collate_fn=custom_collate_fn,
    num_workers=0,
    pin_memory=False,
    drop_last=False
)
```

```python
# Input waveform shapes
waveforms, arrival_times = next(iter(loader))
print(waveforms.shape)
ticks = np.arange(waveforms[0].shape[1])

for wf in waveforms:
    plt.plot(ticks, wf.squeeze(0).cpu(), alpha=0.5)
plt.xlabel(f"Time Tick, 1 tick = 0.001 us")
plt.ylabel("Photon Count")
plt.title(f"Batch of {len(waveforms)} 1D Waveforms")
```

```python
# Inspecting arrival times (gt data)
print(arrival_times[0].shape)
print(arrival_times[0][:10])

idx = 5
wf = waveforms[idx]
plt.plot(ticks, wf.squeeze(0).cpu())

arrival_vector = arrival_times[idx].squeeze(0)
arrival_index = (arrival_vector == 1.0).nonzero(as_tuple=True)[0].item()
print(arrival_index)
print(f"adc value: {wf[0, arrival_index]}")

plt.axvline(x=arrival_index, color='r', linestyle='--', label='True arrival')

plt.legend()
plt.xlabel("Time")
plt.ylabel("Waveform amplitude")
plt.title("Waveform with true arrival time")
plt.show()
```

```python
# Inspecting output of model
# Doing the training yuhh
from model import *
from utils import *

epochs = 1
device = 'cuda'

model = UNet1D()
model.to(device)

optimizer = optim.Adam(model.parameters(), lr=1e-4, weight_decay=1e-6)
train_progress = tqdm(loader, desc=f"Epoch {1}/{epochs}", leave=False, position=0)
for data, target in train_progress:
    data, target = data.to(device), target.to(device)
    optimizer.zero_grad()
    output = model(data)
    break

print(f"full output shape: {output.shape}")
print(f"one waveform shape: {output[0].shape}")
print(output[0][:10])
```
